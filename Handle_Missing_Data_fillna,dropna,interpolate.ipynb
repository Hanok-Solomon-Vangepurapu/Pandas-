```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Pandas/1weather_data.csv",parse_dates=["day"])
df
```

```python
df.set_index('day',inplace=True)
df
```

```python
newdf=df.fillna(0)
newdf
```

```python
newdf =df.fillna(
    {
        "temperature":0,"windspeed":0,
     "event":"No Event"
    }
)
```

```python
newdf
```

```python
#ffill forward means forward fill,or carry forward
newdf=df.fillna(method="ffill")
newdf
```

```python
#bfill is  backward filling
#newdf=df.fillna(method="bfill")
#newdf
```

```python
newdf=df.fillna(method="ffill",limit=1)
newdf
```

```python
newdf=df.interpolate(method="time")
#interpolate is just like guesing the values or like avg between to values
newdf
```

```python
newdf=df.dropna(how="all")
newdf
```

```python
newdf=df.dropna(thresh=1)
newdf
```
